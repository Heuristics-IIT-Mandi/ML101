```python
q3
```

```python
iqr = q3-q1 #(interquartile range=Difference between 3rd quartile and 1st quartile)
```

```python
iqr
```

```python
u_w = q3 + (1.5 * (iqr))
#upper viscous{If any value is greater then upper viscous then it will be outlier at upper viscous}
```

```python
l_w = q1 - (1.5 * iqr)
#lower viscous{If any value is lower then upper viscous then it will be outlier at upper viscous}
```
